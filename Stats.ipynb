```julia
using StatsBase, Distributions, PyPlot, Histograms
plot(h::ErrorHistogram, kw...) = PyPlot.errorbar(
    Base.midpoints(edges(h)[1]),
    contents(h),
    errors(h), kw...
);
```

```julia
s = 5
b = 50
p1 = Poisson(b)
p2 = Poisson(s+b)
```

```julia
h1 = ErrorHistogram(linspace(0,100,101));
h2 = ErrorHistogram(linspace(0,100,101));
```

```julia
for v in rand(p1, 100000)
    push!(h1, Float64(v))
end
for v in rand(p2, 100000)
    push!(h2, Float64(v))
end
```

```julia
PyPlot.figure(figsize=(4,4))

plot(h1)
plot(h2);
```
